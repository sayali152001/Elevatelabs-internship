```python
#importing libraries
import numpy as np
import pandas as pd
```

```python
#importing csv file
data=pd.read_csv("car_prices.csv")
```

```python
#taking the overview of file
data
```

```python
#checking for null values using isnull
print(data.isnull().sum())
```

```python
#handling null values
data = data.dropna(subset=['vin'])
```

```python
#has 12 missing values
data = data.dropna(subset=['sellingprice', 'saledate'])
```

```python
#has more than 11k missing values
data['condition'] = data['condition'].fillna('unknown')
```

```python
data['color'] = data['color'].fillna('unknown')
data['interior'] = data['interior'].fillna('unknown')
```

```python
#cloumns with most missing values, so we can drop these rows
data = data.dropna(subset=['make', 'model', 'trim'])
```

```python
# 13k values missing
data['body'] = data['body'].fillna('unknown')
```

```python
#majority of values are missing 
data = data.drop(columns=['transmission'])
```

```python
# Confirm missing values are handled
print(data.isnull().sum())
```

```python
data
```

```python
#checking duplicate values
duplicates = data.duplicated()
```

```python
print(data[data.duplicated()])
```

```python
#removing duplicates basedon specific column
data = data.drop_duplicates(subset=['vin', 'saledate'])
```

```python
#final check
print(f"Duplicates remaining: {data.duplicated().sum()}")
```

```python
#standardize column headers
data.columns = [col.strip().capitalize() for col in data.columns]
```
